```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Let's create some synthetic documents
from sklearn.datasets import make_classification
X,y= make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, n_classes=2,n_redundant=0)
X
```

```python
y
```

```python
X_data=pd.DataFrame(X)
```

```python
sns.scatterplot(x=X_data[0],y=X_data[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_data,y,test_size=0.2,random_state=42)
```

**Time to import and implement the support vector classifier model**

```python
from sklearn.svm import SVC
```

```python
svc= SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
